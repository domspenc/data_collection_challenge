### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

```python
# Visit the website
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
table = soup.find('table', class_='table')
```

```python
# Create an empty list

columns = []
data = []

# Loop through the rows

headings = table.find_all('th')
for heading in headings:
    columns.append(heading.get_text())
```

```python
# Extract all rows of data
nasa_data = soup.find("table", class_="table")
nasa_data
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Here’s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list
table_columns = []
table_data = []

# Loop through the scraped data to find and store the headers
# 1. Locate the headings data. 
# 2. Pull each heading using a for loop, and add it to the list.
headings_data = nasa_data.find_all('th')
for items in headings_data:
    table_columns.append(items.get_text())

# Loop through the scraped data to find and store the row data
# 1. Locate the main table data.
# 2. Pull each row of data using a for loop, and add it to the list (two part process).
row_data = nasa_data.find_all('tr', class_='data-row')
for rows in row_data:
    data = []
    each_row = rows.find_all('td')
    for items in each_row:
        data.append(items.text)
    table_data.append(data)
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
nasa_df = pd.DataFrame(table_data, columns = table_columns)
```

```python
# Confirm DataFrame was created successfully
nasa_df
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
nasa_df.dtypes
```

```python
# Change column headings for readability
nasa_df = nasa_df.rename(columns={"sol": "solar_day", "ls": "solar_longitude"})
```

```python
# Change data types for data analysis
nasa_df['terrestrial_date'] = pd.to_datetime(nasa_df['terrestrial_date'])
nasa_df = nasa_df.astype({"id": "int", "solar_day": "int", "solar_longitude": "int", "month": "int", "min_temp": "float", "pressure": "float"})
```

```python
# Confirm type changes were successful by examining data types again
nasa_df.dtypes
```

### Step 5: Analyse the Data

Analyse your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
print(f" The number of months on Mars is: {nasa_df['month'].max()}")
```

```python
# 2. How many Martian days' worth of data are there?
print(f" There is {nasa_df['solar_day'].nunique()} Martian days' worth of data in this dataset.")
```

```python
# 3. What is the average low temperature by month?
mars_low_temp = nasa_df.groupby(['month']).mean(['min_temp'])
mars_low_temp_plot = mars_low_temp[['min_temp']].round(2)
mars_low_temp_plot
```

```python
# Plot the average temperature by month
mars_low_temp_plot.plot(kind="bar", figsize=(6,5),color='g', alpha=0.8, align="center")
plt.title('Average Minimum Temperature on Mars by Martian month')
plt.xlabel("Month")
plt.ylabel("Temperature (C)")
plt.tight_layout()
```

```python
# Identify the coldest and hottest months in Curiosity's location
print(f" The coldest month on Mars is {mars_low_temp_plot['min_temp'].idxmin()}")
print(f" The warmest month on Mars is {mars_low_temp_plot['min_temp'].idxmax()}")
```

```python
# 4. Average pressure by Martian month
mars_avg_pressure = nasa_df.groupby(['month']).mean(['pressure'])
mars_avg_pressure_plot = mars_avg_pressure[['pressure']].round(2)
mars_avg_pressure_plot
```

```python
# Plot the average pressure by month
mars_avg_pressure_plot.plot(kind="bar", figsize=(6,5),color='g', alpha=0.8, align="center")
plt.title('Average Pressure on Mars by Martian month')
plt.xlabel("Month")
plt.ylabel("Temperature (C)")
plt.tight_layout()
```

```python
# Identify the highest and lowest recorded atmospheric pressure in Curiosity's location
print(f" The highest atmospheric pressure recorded on Mars is {mars_avg_pressure_plot['pressure'].idxmax()}")
print(f" The lowest atmospheric pressure recorded on Mars is {mars_avg_pressure_plot['pressure'].idxmin()}")
```

```python
fig, ax = plt.subplots(figsize=[15,7])
ax = plt.plot(nasa_df.solar_day,nasa_df.min_temp)
plt.grid(visible=True, which='both')
plt.xlabel('Solar Days Count')
plt.ylabel('Temperature (C)')
plt.title('Minimum Temperature on Mars over time (Martian days)')
plt.minorticks_on()
plt.show()
```

---
#### 5. How many terrestrial (earth) days are there in a Martian year?Using the above graph, it is possible to estimate the number of terrestrial days by looking at the highest and lowest temperature points.

These points indicate a collection of datapoints taken on different solar days on Mars, with the warmest 'peaks' indicating a Mars 'summer', and the coldest 'dips' indicating a Mars 'winter'. To calculate a full revolution of the sun around Mars, we make the following calculations:</br>

Number of days between first and second 'summers': approx 800 - 150 = 650 </br>
Number of days between first and second 'winters': approx 1175 - 540 = 635 </br>
Number of days between second and third 'summers': approx 1495 - 800 = 695</br>

Average number of days across above timeframe: (650 + 635 + 695) / 3 = <strong>660 days</strong></br>

A quote from an educational video on Nasa's website says:</br>
<em>"The Earth zips around the Sun at about 67,000 miles per hour, making a full revolution in about 365 days – one year on Earth. Mars is a little slower, and farther from the sun, so a full circuit takes <strong>687</strong> Earth days – or one Mars year."</em></br>

So our results indicate a decent estimate!

---

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms!



Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth.

The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
